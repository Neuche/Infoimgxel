```python
# Cargamos imagen del mono
X_2 = imread('monkey.jpeg')
```

```python
diffsX1 = np.subtract(X_1.shape, X_2.shape)
diffsX2 = np.subtract(X_2.shape, X_1.shape)
def cropsI(a):
    return int(np.ceil(max(a/2,0)))
def cropsD(a):
    return int(np.floor(max(a/2,0)))
```

```python
# Ejercicio 4
# Cropear las imagenes para que queden del mismo tamaño y concatenarlas vertical y horizonalmente en una única imagen 
# Mostrar los resultado concatenados
# (Cada concatenación se puede hacer en una sola linea de codigo (Mirar vstack y hstack))

matplotlib.pyplot.imshow(
    np.vstack(
        (X_1[cropsI(diffsX1[0]):X_1.shape[0]-cropsD(diffsX1[0]), cropsI(diffsX1[1]):X_1.shape[1]-cropsD(diffsX1[1]), :],
        X_2[cropsI(diffsX2[0]):X_2.shape[0]-cropsD(diffsX2[0]), cropsI(diffsX2[1]):X_2.shape[1]-cropsD(diffsX2[1]), :])
    )
)
```

```python
matplotlib.pyplot.imshow(
    np.hstack(
        (X_1[cropsI(diffsX1[0]):X_1.shape[0]-cropsD(diffsX1[0]), cropsI(diffsX1[1]):X_1.shape[1]-cropsD(diffsX1[1]), :],
        X_2[cropsI(diffsX2[0]):X_2.shape[0]-cropsD(diffsX2[0]), cropsI(diffsX2[1]):X_2.shape[1]-cropsD(diffsX2[1]), :])
    )
)
```

```python
# Ejercicio 6
# Imprimir los 20 valores de intensidad de rojo que más aparecen en el panda
# ¿Cuantas veces aparece la componente 172?
X_1[:,:,:1]
# Tally occurrences of words in a list
cnt = Counter()
for intensity in X_1[:,:,:1].flatten():
    cnt[intensity] += 1
cnt.most_common(20)
```
